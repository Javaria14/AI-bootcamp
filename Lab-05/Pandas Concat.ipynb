## <font color='blue'>Basic Concatenation</font>

```python
import pandas as pd

pakistani_weather = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
pakistani_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([pakistani_weather, us_weather])
df
```

## <font color='blue'>Ignore Index</font>

```python
df = pd.concat([pakistani_weather, us_weather], ignore_index=True)
df
```

## <font color='blue'>Concatenation And Keys</font>

```python
df = pd.concat([pakistani_weather, us_weather], keys=["pakistan", "us"])
df
```

```python
df.loc["us"]
```

```python
df.loc["pakistan"]
```

## <font color='blue'>Concatenation Using Index</font>

```python
temperature_df = pd.DataFrame({
    "city": ["karachi","lahore","peshawar"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["lahore","karachi"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

## <font color='blue'>Concatenate dataframe with series</font>

```python
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```

```python
#Practice
#Pandas concatenation

import pandas as pd

# Create sample DataFrames
df1 = pd.DataFrame({
    "Name": ["Ali", "Sara"],
    "Marks": [80, 90]
})

df2 = pd.DataFrame({
    "Name": ["Ahmed", "Ayesha"],
    "Marks": [85, 95]
})

# Concatenate rows (default axis=0)
print("Rows Combined:")
print(pd.concat([df1, df2]))

# Concatenate rows and reset index
print("Rows Combined with New Index:")
print(pd.concat([df1, df2], ignore_index=True))

# Concatenate columns (axis=1)
grades = pd.DataFrame({"Grade": ["A", "A+"]})
print("Columns Combined:")
print(pd.concat([df1, grades], axis=1))

# Add keys to identify each DataFrame
print("Using Keys:")
result = pd.concat([df1, df2], keys=["Class A", "Class B"])
print(result)

# Access data using a key
print("Data of Class A:")
print(result.loc["Class A"])

# Concatenate a Series as a new column
subjects = pd.Series(["Math", "Science"], name="Subject")
print("Add Series as Column:")
print(pd.concat([df1, subjects], axis=1))

# Outer join (keep all indexes)
df3 = pd.DataFrame({"Age": [20, 21]}, index=[1, 2])
print("Outer Join:")
print(pd.concat([df1, df3], axis=1))

# Inner join (keep common indexes only)
print("Inner Join:")
print(pd.concat([df1, df3], axis=1, join="inner"))

# Verify duplicate indexes
try:
    print("Verify Integrity:")
    print(pd.concat([df1, df2], verify_integrity=True))
except ValueError as e:
    print("Error:", e)

# Concatenate multiple DataFrames
df4 = pd.DataFrame({
    "Name": ["Bilal"],
    "Marks": [88]
})

print("Multiple DataFrames:")
print(pd.concat([df1, df2, df4], ignore_index=True))
```
